```python
import os
import shutil

from ase.io import read
```

```python
xyz_path = "/Users/dario/datasets/GWFit/general"
data_path = "/Users/dario/datasets/GWFit/results/TZ3P/general/eqp"
```

```python
data_files = os.listdir(data_path)
num_samples = 0
for data_file in data_files:
    mol = data_file[:-4]
    atoms = read(f"{xyz_path}/{mol}.xyz", format="xyz")
    elems = atoms.get_chemical_symbols()
    num_heavy_elems = 0
    for elem in elems:
        if elem not in ["H", "C", "N", "O", "F"]:
            num_heavy_elems += 1
    if num_heavy_elems == 0:
        shutil.copyfile(f"{xyz_path}/{mol}.xyz", f"/Users/dario/ViSNetGW/test_datasets/Elements/mols/{mol}.xyz")
        shutil.copyfile(f"{data_path}/{data_file}", f"/Users/dario/ViSNetGW/test_datasets/Elements/E_qp/{data_file}")
        num_samples += 1
print(num_samples)
```
