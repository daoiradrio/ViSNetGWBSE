```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics_file = "../visnet_logs/metrics.csv"
#metrics_file = "/Volumes/LaCie/trained_models/ViSNet/try_high_lr_and_schedule/prehomo1M128502532/metrics.csv"
```

```python
col_train = "train_mae"
col_valid = "val_mae"
col_lr = "lr"
```

```python
df = pd.read_csv(metrics_file)

df_train = df[["epoch", col_train]].dropna()
df_valid = df[["epoch", col_valid]].dropna()
df_lr = df[["epoch", col_lr]].dropna()

df_train = df_train.groupby(["epoch"]).mean()
df_valid = df_valid.groupby(["epoch"]).mean()
#df_lr = df_lr.groupby(["epoch"]).mean()

#print(df_train)
print(df_valid)
#print(df_lr)
```

```python
train_loss = df_train[col_train].to_list()
valid_loss = df_valid[col_valid].to_list()

train_loss = train_loss[1:]
valid_loss = valid_loss[1:]

plt.plot([i+1 for i, _ in enumerate(train_loss)], train_loss)
plt.plot([i+1 for i, _ in enumerate(valid_loss)], valid_loss)
plt.show()
```

```python
metrics_files = [
    "../visnet_logs_1M/metrics.csv",
    "../visnet_logs_5M/metrics.csv",
    "../visnet_logs_1M_transfer/metrics.csv",
    "../visnet_logs_5M_transfer/metrics.csv",
]

for metrics_file in metrics_files:
    label_prefix = os.path.basename(os.path.dirname(metrics_file))

    df = pd.read_csv(metrics_file)

    df_train = df[["epoch", col_train]].dropna()
    df_valid = df[["epoch", col_valid]].dropna()

    df_train = df_train.groupby("epoch").mean()
    df_valid = df_valid.groupby("epoch").mean()

    train_loss = df_train[col_train].tolist()[1:]
    valid_loss = df_valid[col_valid].tolist()[1:]

    epochs = [i + 1 for i in range(len(train_loss))]

    #plt.plot(epochs, train_loss, label=f"{label_prefix} - Train")
    plt.plot(epochs, valid_loss, label=f"{label_prefix} - Valid")

plt.xlabel("Epoch")
plt.ylabel("Loss (MAE)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
